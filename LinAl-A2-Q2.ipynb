```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sol(A,b):
    At = np.transpose(A)
    Atb = np.matmul(At,b)
    AtAI = np.linalg.inv(np.matmul(At,A))
    x = np.matmul(AtAI,Atb)
    y = np.matmul(A,x)
    plt.plot(y,'r*',label='Fit')
    plt.plot(b,'b',label='Original')
    plt.legend()
    plt.show()
    m = np.mean(b)
    R = np.matmul(np.transpose(y-m),y-m)/np.matmul(np.transpose(b-m),b-m)
    print("R value = ",np.sqrt(R))
```

```python
M = 10 
N = 10
x = np.linspace(0, 0.001, M)
x = np.reshape(x, (M,1))
```

```python
mu = np.linspace(0, 0.001, N)
A_s = np.ones((M,N))
A_p = np.ones((M,N))
for i in range(N):
    A_s[:,i] = np.reshape(1/(1+np.exp(10000*(mu[i]-x))), (M,))
    A_p[:,i] = np.reshape(np.power(x,i), (M,))
```

```python
b = np.sin(2000*np.pi*x) + np.reshape(np.random.normal(0.0, 0.3, M) ,(M,1))
```

```python
sol(A_s,b)
```

```python
sol(A_p,b)
```
