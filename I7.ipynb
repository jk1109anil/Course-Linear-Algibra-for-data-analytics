```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
import math
```

```python
def f(u,m,n,e):
    k = math.ceil(4*math.log(m)/(e*e/2-e*e*e/3))
    R1 = np.random.normal(0, 1, k*n)
    R = np.reshape(R1, (k,n))/np.sqrt(k)
    v = np.matmul(R, u)
    result_dot = np.matmul(np.transpose(v), v)
    index = 0
    t = math.floor(m*(m-1)/2)
    dot = np.zeros(t)
    mag = np.zeros(t)
    for i in range(m):
        for j in range(i):
            dot[index] = result_dot[i,j]
            a = v[:,i]-v[:,j]
            b = np.reshape(a, (k,1))
            mag[index] = np.sqrt(np.matmul(np.transpose(b), b) )
            index = index+1
    return dot, mag
```

```python
f = open('test.dat', "r")
mat =  np.genfromtxt(f)
u = np.transpose(mat)
[n, m] = np.shape(u)
```

```python
result_dot = np.matmul(np.transpose(u), u)
index = 0
t = math.floor(m*(m-1)/2)
dot = np.zeros(t)
mag = np.zeros(t)
for i in range(m):
    for j in range(i):
        dot[index] = result_dot[i,j]
        a = u[:,i]-u[:,j]
        b = np.reshape(a, (n,1))
        mag[index] = np.sqrt(np.matmul(np.transpose(b), b) )
        index = index+1
```

```python
dot1, mag1 = f(u,m,n,0.5)
dot2, mag2 = f(u,m,n,0.2)
```

```python
line = np.arange(5000)
plt.scatter(dot1, dot)
plt.plot(line,'r')
plt.xlabel('pairwise dot-product (epsilon = 0.5)')
plt.ylabel('pairwise dot-product')
plt.show()
```

```python
plt.scatter(dot2, dot)
plt.plot(line,'r')
plt.xlabel('pairwise dot-product (epsilon = 0.2)')
plt.ylabel('pairwise dot-product')
plt.show()
```

```python
line = np.arange(50)
plt.scatter(mag1, mag)
plt.plot(line,'r')
plt.xlabel('pairwise distance (epsilon = 0.5)')
plt.ylabel('pairwise distance')
plt.show()
```

```python
plt.scatter(mag2, mag)
plt.plot(line,'r')
plt.xlabel('pairwise distance (epsilon = 0.2)')
plt.ylabel('pairwise distance')
plt.show()
```
