```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sol(A,b):
    At = np.transpose(A)
    Atb = np.matmul(At,b)
    AtAI = np.linalg.inv(np.matmul(At,A))
    x = np.matmul(AtAI,Atb)
    y = np.matmul(A,x)
    plt.plot(y,'r*',label='Fit')
    plt.plot(b,'b',label='Original')
    plt.legend()
    plt.show()
    m = np.mean(b)
    R = np.matmul(np.transpose(y-m),y-m)/np.matmul(np.transpose(b-m),b-m)
    print("R value = ",np.sqrt(R))
```

```python
f = open('data3.txt', "r")
mat1 =  np.genfromtxt(f)     # features and labels
```

```python
N = 200
mat = mat1[:N,:]
A = np.ones((N, 2))
```

```python
A[:,:-1] = mat[:,:-1]
b = mat[:,1]
```

```python
sol(A,b)
```
